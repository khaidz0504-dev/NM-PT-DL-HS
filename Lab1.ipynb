```python
df = pd.read_csv("dulieuxettuyendaihoc.csv", index_col="STT")
print("Số giá trị thiếu của DT:", df["DT"].isnull().sum())
df["DT"] = df["DT"].fillna(0)
print("Sau xử lý:", df["DT"].isnull().sum())
```

```python
df["T1"].fillna(df["T1"].mean(), inplace=True)
df
```

```python
df = df.fillna(df.mean(numeric_only=True))
print(df.isnull().sum())
```

```python
df["TBM1"] = (df["T1"]*2 + df["L1"] + df["H1"] + df["S1"] + df["V1"]*2 + df["X1"] + df["D1"] + df["N1"]) / 10
df["TBM2"] = (df["T2"]*2 + df["L2"] + df["H2"] + df["S2"] + df["V2"]*2 + df["X2"] + df["D2"] + df["N2"]) / 10
df["TBM3"] = (df["T3"]*2 + df["L3"] + df["H3"] + df["S3"] + df["V3"]*2 + df["X3"] + df["D3"] + df["N3"]) / 10
df[["TBM1", "TBM2", "TBM3"]].head()
```

```python
import numpy as np
def xep_loai(tbm):
    return np.select(
        [
            tbm < 5,
            tbm < 6.5,
            tbm < 8,
            tbm < 9,
            tbm >= 9
        ],
        ["Y", "TB", "K", "G", "XS"],
        default=''
    )

df["XL1"] = xep_loai(df["TBM1"])
df["XL2"] = xep_loai(df["TBM2"])
df["XL3"] = xep_loai(df["TBM3"])
df[["TBM1", "XL1", "TBM2", "XL2", "TBM3", "XL3"]].head()
```

```python
df["US_TBM1"] = (df["TBM1"] - 0) / (10 - 0) * 4
df["US_TBM2"] = (df["TBM2"] - 0) / (10 - 0) * 4
df["US_TBM3"] = (df["TBM3"] - 0) / (10 - 0) * 4
df[["TBM1", "US_TBM1", "TBM2", "US_TBM2", "TBM3", "US_TBM3"]].head()
```

```python
def tinh_kqxt(row):
    if row["KT"] in ["A", "A1"]:
        diem = (row["DH1"]*2 + row["DH2"] + row["DH3"]) / 4
    elif row["KT"] == "B":
        diem = (row["DH1"] + row["DH2"]*2 + row["DH3"]) / 4
    else:
        diem = (row["DH1"] + row["DH2"] + row["DH3"]) / 3

    return 1 if diem >= 5 else 0

df["KQXT"] = df.apply(tinh_kqxt, axis=1)
df[["KT", "DH1", "DH2", "DH3", "KQXT"]].head(10)
```

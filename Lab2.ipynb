```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('processed_dulieuxettuyendaihoc.csv')
df.head()
```

```python
pivot1 = pd.pivot_table(
    df,
    values='DH1',
    index='KT',
    aggfunc=[
        'count',
        'sum',
        'mean',
        'median',
        'min',
        'max',
        'std'
    ]
)

pivot1
```

```python
pivot2 = pd.pivot_table(
    df,
    values='DH1',
    index=['KT','KV'],
    aggfunc=[
        'count',
        'sum',
        'mean',
        'median',
        'min',
        'max',
        'std'
    ]
)

q = df.groupby(['KT','KV'])['DH1'].quantile([0.25,0.5,0.75]).unstack()
q.columns = pd.MultiIndex.from_product([['DH1'], ['Q1','Q2','Q3']])

pivot2.join(q)
```

```python
pivot3 = pd.pivot_table(
    df,
    values='DH1',
    index=['KT','KV','DT'],
    aggfunc=[
        'count',
        'sum',
        'mean',
        'median',
        'min',
        'max',
        'std'
    ]
)

q = df.groupby(['KT','KV','DT'])['DH1'].quantile([0.25,0.5,0.75]).unstack()
q.columns = pd.MultiIndex.from_product([['DH1'], ['Q1','Q2','Q3']])

pivot3.join(q)
```

```python
# Câu 4 + Câu 5

# Create 'phanlopT1' column by categorizing 'T1' scores
bins = [0, 5, 7, 8.5, 10]
labels = ['Yếu', 'TB', 'Khá', 'Giỏi']
df['phanlopT1'] = pd.cut(df['T1'], bins=bins, labels=labels, right=False)

plt.figure(figsize=(14,6))

# Multiple Line
for ten, nhom in df.groupby('phanlopT1'):
    plt.plot(
        nhom.index,
        nhom['T1'],
        marker='o',
        linewidth=1,
        label=ten
    )

# Drop-line
plt.vlines(
    x=df.index,
    ymin=0,
    ymax=df['T1'],
    colors='gray',
    alpha=0.3
)

plt.title("Biểu đồ Multiple Line kết hợp Drop-line cho T1")
plt.xlabel("Học sinh")
plt.ylabel("Điểm T1")
plt.legend(title="Phân lớp")
plt.grid(True)

plt.show()
```

```python
# ==========================
# PHẦN 5
# ==========================

import scipy.stats as stats
import statsmodels.api as sm

# --------------------------
# Câu 1
# --------------------------

print("===== MÔ TẢ T1 =====")
print(df['T1'].describe())

print("\nSkewness:", df['T1'].skew())
print("Kurtosis:", df['T1'].kurt())

plt.figure(figsize=(6,4))
plt.boxplot(df['T1'], vert=False)
plt.title("Boxplot T1")
plt.show()

plt.figure(figsize=(6,4))
plt.hist(df['T1'], bins=15)
plt.title("Histogram T1")
plt.show()

sm.qqplot(df['T1'], line='45')
plt.title("QQ Plot T1")
plt.show()

# --------------------------
# Câu 2
# --------------------------

df.boxplot(column='T1', by='phanlopT1')
plt.suptitle("")
plt.title("Boxplot theo phân lớp")
plt.show()

for ten, nhom in df.groupby("phanlopT1"):

    plt.figure(figsize=(6,4))
    plt.hist(nhom["T1"], bins=10)
    plt.title(f"Histogram {ten}")
    plt.show()

    sm.qqplot(nhom["T1"], line='45')
    plt.title(f"QQ Plot {ten}")
    plt.show()

# --------------------------
# Câu 3
# --------------------------

print("\n===== DH1 và T1 =====")

print("Covariance")
print(df[['DH1','T1']].cov())

print("\nCorrelation")
print(df[['DH1','T1']].corr())

plt.figure(figsize=(6,4))
plt.scatter(df["T1"],df["DH1"])
plt.xlabel("T1")
plt.ylabel("DH1")
plt.title("Scatter DH1 - T1")
plt.show()

# --------------------------
# Câu 4
# --------------------------

for kv, nhom in df.groupby("KV"):

    plt.figure(figsize=(6,4))

    plt.scatter(
        nhom["T1"],
        nhom["DH1"]
    )

    plt.title(f"Khu vực {kv}")
    plt.xlabel("T1")
    plt.ylabel("DH1")

    plt.show()

# --------------------------
# Câu 5
# --------------------------

print("\n===== DH1 DH2 DH3 =====")

print("Covariance")
print(df[['DH1','DH2','DH3']].cov())

print("\nCorrelation")
print(df[['DH1','DH2','DH3']].corr())

plt.figure(figsize=(6,4))
plt.scatter(df["DH1"],df["DH2"])
plt.xlabel("DH1")
plt.ylabel("DH2")
plt.title("DH1 - DH2")
plt.show()

plt.figure(figsize=(6,4))
plt.scatter(df["DH1"],df["DH3"])
plt.xlabel("DH1")
plt.ylabel("DH3")
plt.title("DH1 - DH3")
plt.show()

plt.figure(figsize=(6,4))
plt.scatter(df["DH2"],df["DH3"])
plt.xlabel("DH2")
plt.ylabel("DH3")
plt.title("DH2 - DH3")
plt.show()
```
